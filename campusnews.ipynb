```python
import pandas as pd
```

```python
wikiurl = "https://en.wikipedia.org/wiki/List_of_college_and_university_student_newspapers_in_the_United_States"
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
wikidump = requests.get(wikiurl)
```

```python
wikibs = BeautifulSoup(wikidump.content, "html.parser")
```

```python
wikibs
```

```python
campusnews = pd.read_csv("data/campusnews.csv")
```

```python
campusnews.columns
```

```python
urlcols = [col for col in campusnews.columns if "URL" in col]
```

```python
from itertools import chain
```

```python
urls = set(list(chain(*[campusnews[col] for col in urlcols])))
```

```python
urls = [url for url in urls if type(url) == str]
```

```python
urls
```

```python
wiki_a = wikibs.find_all("a", href = True)
```

```python
wiki_hrefs = [item["href"] for item in wiki_a]
```

```python
wiki_hrefs
```

```python
wiki_i = wikibs.find_all("i")
```

```python
wiki_a = [item.find_all("a", href = True) for item in wiki_i]
```

```python
wiki_a = [l for l in wiki_a if len(l) > 0]
```

```python
wiki_a = [l[0] for l in wiki_a]
```

```python
wiki_a
```

```python
wiki_hrefs = [item["href"] for item in wiki_a]
```

```python
wiki_hrefs = [h for h in wiki_hrefs if "redlink" not in h and "template_removal" not in h]
```

```python
wiki_hrefs
```

```python
wiki_hrefs = [f"https://en.wikipedia.org/{relurl}" for relurl in wiki_hrefs]
```

```python
wiki_hrefs
```

```python
example_wiki = requests.get('https://en.wikipedia.org//wiki/The_Auburn_Plainsman')
```

```python
wiki_example = BeautifulSoup(example_wiki.content, "html.parser")
```

```python
wiki_example
```

```python
paperlink = [item.find_all("a", class_= "external text") for item in wiki_example.find_all("table")]
```

```python
paperlink
```

```python
wiki_papers = []
```

```python
import time
```

```python
for url in wiki_hrefs:
    rq = requests.get(url)
    bs = BeautifulSoup(rq.content, "html.parser")
    links = [item.find_all("a", class_ = "external text") for item in bs.find_all("table")]
    links = list(chain(*[item for item in links if len(item) > 0]))
    finallink = [link["href"] for link in links if "worldcat" not in link["href"] and "google.com" not in link["href"] and "geohack" not in link["href"] and "jstor" not in link["href"] and "wikipedia" not in link["href"]]
    if len(finallink) > 0:
        wiki_papers.append(finallink[0])
        print(finallink[0])
    time.sleep(1)
```

```python
df = pd.DataFrame(data = urls + wiki_papers, columns = ["URL"])
```

```python
df
```

```python
df.to_csv("campus_urls.csv")
```
